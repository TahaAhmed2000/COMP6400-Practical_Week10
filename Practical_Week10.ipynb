# Task 1 (b)

Given the two list of numbers:

list_1 = [2.0,3.5,4.12,5.65,6.11,7.77,8.1], and

list_2 = [10,8,9,6.13,14,15]

1. Find the mean, max and min of each list (use the NumPy library)
2. Add two numbers to each list using .append() method; add 1.5, 2.7 to **list_1** and 20, 25 to **list_2**
3. Find the mean of each updated list
3. Concatenate both the lists and sort the resultant list in ascending order (use the NumPy library)

```python
# write your code here:
list_1 = [2.0,3.5,4.12,5.65,6.11,7.77,8.1]
list_2 = [10,8,9,6.13,14,15]

import numpy as np

print(np.mean(list_1), np.max(list_1),np.min(list_1))
print(np.mean(list_2), np.max(list_2),np.min(list_2))

list_1.append(1.5)
list_1.append(2.7)
list_2.append(20)
list_2.append(25)
print('\n',list_1)
print(np.mean(list_1))
print('\n',list_2)
print(np.mean(list_2))

list_3=list_1+list_2
print(np.sort(list_3))
```

# Task 2: Function, Loop and Conditional statements

Write a function that will take **list_1** as input, and return the count of numbers that are strictly greater than 5. 

```python
# Function definition

def count_num(input_list):
  # write your code here
    count=0
    if len(input_list)>5:
        for i in range(len(input_list)):
            count+=1
    return count
```

```python
# Function call

count_num(list_1)
```
